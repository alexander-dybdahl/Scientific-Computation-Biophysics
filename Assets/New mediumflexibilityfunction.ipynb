```python
##Oppgave 2e)

def mediumFlexibilityMove(k, direction, grid):
    N = len(grid)
    newGrid = copy.copy(grid)
    isolatedGrid = np.zeros([N,N])
    locationOfMonomers = np.transpose(np.nonzero(grid))

    for y, x in locationOfMonomers:
        if grid[y, x] == k:
            isolatedGrid[y, x] = k


    movedGrid = list()
    emptyList = np.zeros(N)

    if direction == 0: # right
        for i in range(N):
            row = grid[i]
            choosenMonomers = list()
            allowedMove = True
            for j in range(N):
                if row[j] == k:
                    choosenMonomers.append([i,j])

                if row[j] != k and row[j] != 0 and row[(j-1)%N] == k:
                    allowedMove = False

            if allowedMove == True:
                for y, x in choosenMonomers:
                    newGrid[y, x] = 0
                movedGrid.append(isolatedGrid[i])
            else:
                movedGrid.append(emptyList.tolist())
        
        newIsolatedGrid = np.roll(movedGrid,1,1)


    if direction == 1: # left
        for i in range(len(grid)):
            row = grid[i]
            choosenMonomers = list()
            allowedMove = True
            for j in range(N):
                if row[-1-j] == k:
                    choosenMonomers.append([i,-1-j%N])

                if row[-1-j] != k and row[-1-j] != 0 and row[-1-(j-1)%N] == k:
                    allowedMove = False

            if allowedMove == True:
                for y, x in choosenMonomers:
                    newGrid[y, x] = 0
                movedGrid.append(isolatedGrid[i])
            else:
                movedGrid.append(emptyList.tolist())

        newIsolatedGrid = np.roll(movedGrid,-1,1)

    if direction == 2: # up
        gridT = np.transpose(grid)
        newGridT = copy.copy(gridT)
        isolatedGridT = np.transpose(isolatedGrid)

        for i in range(N):
            row = gridT[i]
            choosenMonomers = list()
            allowedMove = True
            for j in range(N):
                if row[j] == k:
                    choosenMonomers.append([i,j%N])

                if row[j] != k and row[j] != 0 and row[(j+1)%N] == k:
                    allowedMove = False

            if allowedMove == True:
                for y, x in choosenMonomers:
                    newGridT[y, x] = 0
                movedGrid.append(isolatedGridT[i])
            else:
                movedGrid.append(emptyList.tolist())

        newGrid = np.transpose(newGridT)
        movedGrid = np.transpose(movedGrid)
        newIsolatedGrid = np.roll(movedGrid,-1,0)


    if direction == 3: # down
        gridT = np.transpose(grid)
        newGridT = copy.copy(gridT)
        isolatedGridT = np.transpose(isolatedGrid)

        for i in range(N):
            row = gridT[i]
            choosenMonomers = list()
            allowedMove = True
            for j in range(N):
                if row[-1-j] == k:
                    choosenMonomers.append([i,-1-j%N])

                if row[-1-j] != k and row[-1-j] != 0 and row[-1-(j+1)%N] == k:
                    allowedMove = False

            if allowedMove == True:
                for y, x in choosenMonomers:
                    newGridT[y, x] = 0
                movedGrid.append(isolatedGridT[i])
            else:
                movedGrid.append(emptyList.tolist())

        newGrid = np.transpose(newGridT)
        movedGrid = np.transpose(movedGrid)
        newIsolatedGrid = np.roll(movedGrid,1,0)

    if allowedMove == True:
        return newGrid + newIsolatedGrid
    else:
        return grid


N = 20
M = 2
L = 20

grid = createSystemOfPolymers(N, M, L)

k = -2
direction = 3

movedGrid = mediumFlexibilityMove(k, direction, grid)

plotSystem(grid, "Grid before movement", False, True)
plotSystem(movedGrid, "Medium flexibility movement grid", False, True)
```
